1. K-Means Clustering[SUB OPTIMAL]:
- Load customer data.
- Check the number of cells in each column with null values.
- Create a scatter plot with Age as X and Spending Score as Y.
- Draw a scatter plot displaying data points colored on the basis of clusters

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
df=pd.read_csv(r"G:\Intellipaat-20220522T063246Z-001\Intellipaat Assignments\Unsupervised Learning Assignments\1.-KMeans-Suboptimal\customers.csv")
df.head()
```

```python
sns.scatterplot(x=df['Age'],y=df['Spending Score (1-100)'],color='b')
```

```python
X=df.iloc[:,2:]
X.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
clusters = kmeans.predict(X)
df['label'] = clusters
df.head()
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
#3D Plot as we did the clustering on the basis of 3 input features
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.Age[df.label == 0], df["Annual Income (k$)"][df.label == 0], df["Spending Score (1-100)"][df.label == 0], c='purple', s=60)
ax.scatter(df.Age[df.label == 1], df["Annual Income (k$)"][df.label == 1], df["Spending Score (1-100)"][df.label == 1], c='red', s=60)
ax.scatter(df.Age[df.label == 2], df["Annual Income (k$)"][df.label == 2], df["Spending Score (1-100)"][df.label == 2], c='blue', s=60)
ax.view_init(35, 185)
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```
